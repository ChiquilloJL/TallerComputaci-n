```python
# Importando todas las librerías requeridas para el análisis

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Cargando el archivo commo dataframe
df_dvio = pd.read_csv("./data/violencia.csv")
```

```python
# Nombre y tipo de datos de las columnas
df_dvio.dtypes
```

```python
# Cantidad y porcentaje de cada departamento de los casos reportados como Violnecia Intrafamiliar
abs_frecuency = df_dvio["DEPARTAMENTO"].value_counts().to_frame()
rel_frecuency = df_dvio["DEPARTAMENTO"].value_counts(normalize = True).to_frame()
pd.concat([abs_frecuency, rel_frecuency], axis = 1)
```

```python
#TOP 3 departamentos más afectados por violencia intrafamiliar de Enero de 2010 hasta Julio de 2021
abs_frecuency.head(3)
```

```python
#Convertir el set FECHA HECHO de type object a type date
anio = pd.to_datetime(df_dvio["FECHA HECHO"])
```

```python
#TOP 3 departamentos más afectados por violencia intrafamiliar en el último año
df_ultanio = df_dvio.loc[anio > "2020-07-31"]
absanio_frecuency = df_ultanio["DEPARTAMENTO"].value_counts().to_frame()
absanio_frecuency.head(3)
```

```python
#Arma más común en los delitos de violencia intrafamiliar.
arm_frecuency = df_dvio["ARMAS MEDIOS"].value_counts().to_frame()
arm_frecuency.head(1)
```

```python
#Porcentaje y cantidad de cada género que es reportado como victíma de Violencia intrafamiliar
rel_frecuencyG = df_dvio["GENERO"].value_counts(normalize = True).to_frame()
abs_frecuencyG = df_dvio["GENERO"].value_counts().to_frame()
pd.concat([abs_frecuencyG, rel_frecuencyG], axis = 1)
```

```python
#Porcentaje y cantidad del último año por género reportado como Violencia intrafamiliar
df_ultanioG = df_dvio.loc[anio > "2020-07-31"]
relanio_frecuencyG = df_ultanioG["GENERO"].value_counts(normalize = True).to_frame()
absanio_frecuencyG = df_ultanioG["GENERO"].value_counts().to_frame()
pd.concat([absanio_frecuencyG, relanio_frecuencyG], axis = 1)
```

```python
#Porcentaje y cantidad de los grupos etarios victímas de violencia intrafamiliar
rel_frecuencyGr = df_dvio["GRUPO ETARIO"].value_counts(normalize = True).to_frame()
abs_frecuencyGr = df_dvio["GRUPO ETARIO"].value_counts().to_frame()
pd.concat([abs_frecuencyGr, rel_frecuencyGr], axis = 1)
```

```python
#Porcentaje y cantidad de los grupos etarios victímas en el último año de violencia intrafamiliar
df_ultanioGr = df_dvio.loc[anio > "2020-07-31"]
relanio_frecuencyGr = df_ultanioGr["GRUPO ETARIO"].value_counts(normalize = True).to_frame()
absanio_frecuencyGr = df_ultanioGr["GRUPO ETARIO"].value_counts().to_frame()
pd.concat([absanio_frecuencyGr, relanio_frecuencyGr], axis = 1)
```
