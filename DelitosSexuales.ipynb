```python
# Importando todas las librerías requeridas para el análisis

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Cargando el archivo commo dataframe
df_dsex = pd.read_csv("./data/sexuales.csv")
```

```python
# Nombre y tipo de datos de las columnas
df_dsex.dtypes
```

```python
# Cantidad y porcentaje de cada departamento de los casos reportados como delitos sexuales
rel_frecuency = df_dsex["DEPARTAMENTO"].value_counts(normalize = True).to_frame()
abs_frecuency = df_dsex["DEPARTAMENTO"].value_counts().to_frame()
pd.concat([abs_frecuency, rel_frecuency], axis = 1)
```

```python
#TOP 3 departamentos más afectados por delitos sexuales de Enero de 2010 hasta Julio de 2021
abs_frecuency.head(3)
```

```python
#Convertir el campo FECHA HECHO de type object a type date
anio = pd.to_datetime(df_dsex["FECHA HECHO"])
```

```python
#TOP 3 departamentos más afectados por delitos sexuales en el último año
df_ultanio = df_dsex.loc[anio > "2020-07-31"]
absanio_frecuency = df_ultanio["DEPARTAMENTO"].value_counts().to_frame()
absanio_frecuency.head(3)
```

```python
#Medio más común para cometer delitos sexuales.
arm_frecuency = df_dsex["ARMAS MEDIOS"].value_counts().to_frame()
arm_frecuency.head(1)
```

```python
#Porcentaje y cantidad de cada género que es reportado como victíma de delito sexual
rel_frecuencyG = df_dsex["GENERO"].value_counts(normalize = True).to_frame()
abs_frecuencyG = df_dsex["GENERO"].value_counts().to_frame()
pd.concat([abs_frecuencyG, rel_frecuencyG], axis = 1)
```

```python
#Porcentaje y cantidad del último año por género reportado como delito sexual
df_ultanioG = df_dsex.loc[anio > "2020-07-31"]
relanio_frecuencyG = df_ultanioG["GENERO"].value_counts(normalize = True).to_frame()
absanio_frecuencyG = df_ultanioG["GENERO"].value_counts().to_frame()
pd.concat([absanio_frecuencyG, relanio_frecuencyG], axis = 1)
```

```python
#Porcentaje y cantidad de los grupos etarios victímas de los delitos sexuales
rel_frecuencyGr = df_dsex["GRUPO ETARIO"].value_counts(normalize = True).to_frame()
abs_frecuencyGr = df_dsex["GRUPO ETARIO"].value_counts().to_frame()
pd.concat([abs_frecuencyGr, rel_frecuencyGr], axis = 1)
```

```python
#Porcentaje y cantidad de los grupos etarios victímas en el último año de delitos sexuales
df_ultanioGr = df_dsex.loc[anio > "2020-07-31"]
relanio_frecuencyGr = df_ultanioGr["GRUPO ETARIO"].value_counts(normalize = True).to_frame()
absanio_frecuencyGr = df_ultanioGr["GRUPO ETARIO"].value_counts().to_frame()
pd.concat([absanio_frecuencyGr, relanio_frecuencyGr], axis = 1)
```

```python
#Porcentaje y cantidad de los 3 delitos más comunes para delitos sexuales
rel_frecuencyD = df_dsex["delito"].value_counts(normalize = True).to_frame()
abs_frecuencyD = df_dsex["delito"].value_counts().to_frame()
pd.concat([abs_frecuencyD, rel_frecuencyD], axis = 1).head(3)
```
