```python
# Importando todas las librerías requeridas para el análisis

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Cargando el archivo commo dataframe
df_dhom = pd.read_csv("./data/homicidio.csv")
```

```python
# Nombre y tipo de datos de las columnas
df_dhom.dtypes
```

```python
# Cantidad y porcentaje de cada departamento de los casos reportados como delitos de Homicidio
abs_frecuency = df_dhom["DEPARTAMENTO"].value_counts().to_frame()
rel_frecuency = df_dhom["DEPARTAMENTO"].value_counts(normalize = True).to_frame()
pd.concat([abs_frecuency, rel_frecuency], axis = 1)
```

```python
#TOP 3 departamentos más afectados por delitos de homicidio de Enero de 2010 hasta Julio de 2021
abs_frecuency.head(3)
```

```python
#Convertir el set FECHA HECHO de type object a type date
anio = pd.to_datetime(df_dhom["FECHA HECHO"])
```

```python
#TOP 3 departamentos más afectados por delitos de homicidios en el último año
df_ultanio = df_dhom.loc[anio > "2020-07-31"]
absanio_frecuency = df_ultanio["DEPARTAMENTO"].value_counts().to_frame()
absanio_frecuency.head(3)
```

```python
#Arma más común usada para delitos de homicidio.
arm_frecuency = df_dhom["ARMAS MEDIOS"].value_counts().to_frame()
arm_frecuency.head(1)
```

```python
#Porcentaje y cantidad de cada género que es reportado como victíma de Homicidio
rel_frecuencyG = df_dhom["GENERO"].value_counts(normalize = True).to_frame()
abs_frecuencyG = df_dhom["GENERO"].value_counts().to_frame()
pd.concat([abs_frecuencyG, rel_frecuencyG], axis = 1)
#Se reporta que aparecen campos vacíos en el data set usado. En los campos de genero. 
```

```python
#Porcentaje y cantidad del último año por género reportado como Homicidio
df_ultanioG = df_dhom.loc[anio > "2020-07-31"]
relanio_frecuencyG = df_ultanioG["GENERO"].value_counts(normalize = True).to_frame()
absanio_frecuencyG = df_ultanioG["GENERO"].value_counts().to_frame()
pd.concat([absanio_frecuencyG, relanio_frecuencyG], axis = 1)
```

```python
#Porcentaje y cantidad de los grupos etarios victímas de homicidios
rel_frecuencyGr = df_dhom["GRUPO ETARÍO"].value_counts(normalize = True).to_frame()
abs_frecuencyGr = df_dhom["GRUPO ETARÍO"].value_counts().to_frame()
pd.concat([abs_frecuencyGr, rel_frecuencyGr], axis = 1)
```

```python
#Porcentaje y cantidad de los grupos etarios victímas en el último año de homicidios
df_ultanioGr = df_dhom.loc[anio > "2020-07-31"]
relanio_frecuencyGr = df_ultanioGr["GRUPO ETARÍO"].value_counts(normalize = True).to_frame()
absanio_frecuencyGr = df_ultanioGr["GRUPO ETARÍO"].value_counts().to_frame()
pd.concat([absanio_frecuencyGr, relanio_frecuencyGr], axis = 1)
```

```python
#Porcentaje y cantidad de los 3 delitos más comunes para Homicidios
rel_frecuencyD = df_dhom["DESCRIPCIÓN CONDUCTA"].value_counts(normalize = True).to_frame()
abs_frecuencyD = df_dhom["DESCRIPCIÓN CONDUCTA"].value_counts().to_frame()
pd.concat([abs_frecuencyD, rel_frecuencyD], axis = 1).head(3)
```
